```python
import os
import numpy as np
import matplotlib.pyplot as plt
import rubin_sim.movingObjects as mo
from rubin_sim.data import get_data_dir
```

```python
ddir = os.path.join(get_data_dir(), 'orbits')
```

```python
trojans = mo.Orbits()
trojans.readOrbits(os.path.join(ddir, 'trojan_5k.txt'))
```

```python
trojans.orbits['a'] = trojans.orbits.q / (1 - trojans.orbits.e)
```

```python
plt.figure(figsize=(8,5))
plt.plot(trojans.orbits.a,  trojans.orbits.e, 'k.')
plt.xlabel('semi-major axis (a) (AU)', fontsize='x-large')
plt.ylabel('eccentricity', fontsize='x-large')
```

```python
plt.figure(figsize=(8,5))
plt.plot(trojans.orbits.a,  trojans.orbits.inc, 'k.')
plt.xlabel('semi-major axis (a) (AU)', fontsize='x-large')
plt.ylabel('inclination (deg)', fontsize='x-large')
```

```python
mbas = mo.Orbits()
mbas.readOrbits(os.path.join(ddir, 'mba_5k.txt'))
mbas.orbits['a'] = mbas.orbits.q / (1 - mbas.orbits.e)
```

```python
plt.figure(figsize=(8,5))
plt.plot(trojans.orbits.a,  trojans.orbits.e, 'k.', label='trojan')
plt.plot(mbas.orbits.a, mbas.orbits.e, 'r.', label='mba')
plt.legend(loc='upper right')
plt.xlabel('semi-major axis (a) (AU)', fontsize='x-large')
plt.ylabel('eccentricity', fontsize='x-large')
```

```python
plt.figure(figsize=(8,5))
plt.plot(trojans.orbits.a,  trojans.orbits.inc, 'k.', label='trojan')
plt.plot(mbas.orbits.a, mbas.orbits.inc, 'r.', label='mba')
plt.legend(loc='upper left')
plt.xlabel('semi-major axis (a) (AU)', fontsize='x-large')
plt.ylabel('inclination (deg)', fontsize='x-large')
```
